```python
from __future__ import division
import operator
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql.functions import col
import numpy as np
sc = SparkContext("local", "Decision Tree")
sqlContext = SQLContext(sc)
```

```python
#importing libraries
import numpy as np
import time
import pandas as pd
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
from pyspark.mllib.util import MLUtils
import math
from pyspark.mllib.regression import LabeledPoint
```

```python
dfinal = pd.read_csv('processed.csv')
dfinal = dfinal.iloc[:,:]
s_dfinal = sqlContext.createDataFrame(dfinal)
type(s_dfinal)
```

```python
data  = s_dfinal.rdd.map(lambda x: LabeledPoint(x[0], x[1:]))
#data = sqlContext.createDataFrame(dfactual).rdd
```

```python
data
```

```python
# Load and parse the data file into an RDD of LabeledPoint.
#data = MLUtils.loadLibSVMFile(sc, 'data/mllib/sample_libsvm_data.txt')
# Split the data into training and test sets (30% held out for testing)

(trainingData, testData) = data.randomSplit([0.7, 0.3])

# Train a DecisionTree model.
#  Empty categoricalFeaturesInfo indicates all features are continuous.
model = DecisionTree.trainClassifier(trainingData, numClasses=2, categoricalFeaturesInfo={},
                                     impurity='gini', maxDepth=30, maxBins=32)

# Evaluate model on test instances and compute test error
predictions = model.predict(testData.map(lambda x: x.features))
labelsAndPredictions = testData.map(lambda lp: lp.label).zip(predictions)
testErr = labelsAndPredictions.filter(
    lambda lp: lp[0] != lp[1]).count() / float(testData.count())
print('Test Error = ' + str(testErr))
print('Learned classification tree model:')
print(model.toDebugString())
```
